```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url="https://www.pngjewellers.com/?srsltid=AfmBOopaYKaf_eKjtykzbOZ2RWZY32knZazkNZMEh70vSUujgIxGFrpe"
```

```python
response = requests.request('get',url).content
```

```python
soup= BeautifulSoup(response,'html.parser')
```

```python

print(soup.prettify())
```

```python
prices=[]
data=soup.find_all('span',class_='text-black inline-block rtl:ml-2 ltr:mr-2')

for price in  data:
    prices.append(price.text.strip())

```

```python
print(prices)
```

```python
import pandas as pd

df_price= pd.DataFrame(prices,columns=['Prices of Jewellery'])
```

```python
df_price
```

```python
# # Extract rows from index 0 to 35 
df_35 = df_price.iloc[:36]
df_35
```

```python
names=[]
data=soup.find_all('p',class_='text-sm')

for name in  data:
    names.append(name.text.strip())
```

```python
print(names)
```

```python
jewleery_names =names[1:43]
```

```python
print(jewleery_names)
```

```python
df_jewllery= pd.DataFrame(jewleery_names,columns=['Jewellery Names'])
```

```python
df_jewllery
```

```python
# # Extract rows from index 0 to 35 
df_till_35 = df_jewllery.iloc[:36]
df_till_35
```

```python
Actor=[]
data=soup.find_all('h6',class_= 'text-sm lg:text-xl text-white font-light')

for model in  data:
    Actor.append(model.text.strip())





```

```python
print(Actor)
```

```python
df_model= pd.DataFrame(Actor,columns=['Brand Ambaseeder'])
```

```python
df_model
```

```python
# Combine DataFrames
df_PNG = pd.concat([df_till_35,df_35,df_model,],axis=1 )
df_PNG
```

```python
df_PNG.to_csv('PNG_data.csv', index=False)
```
